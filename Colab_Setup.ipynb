```python
# Start Flask with Colab public URL
from google.colab.output import eval_js
import subprocess
import time
import threading

print('🚀 Starting Flask server...')

# Start Flask in background thread
def run_flask():
    subprocess.run(['python', 'app_flask.py'])

flask_thread = threading.Thread(target=run_flask, daemon=True)
flask_thread.start()
time.sleep(5)

print('✅ Flask started on port 80!')
print('\n🌐 Access your app at: http://localhost:80')
print('\n💡 Use the URL above in Colab\'s browser or share it!')
```
